```python
##importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load dataset

df = pd.read_csv("EPL_20_21.csv")
```

```python
#i need two more columns for this data to be precised
df["Minspermatch"] = (df["Mins"]/df["Matches"]).astype(int)
df["Goalspermatch"] = (df["Goals"]/df["Matches"]).astype(float)
```

```python
#as you can see two columns had included in this data set
df.tail(20)
```

```python
#total goals for this season
total_goals=df["Goals"].sum()
print(total_goals)
```

```python
#total assists
total_assists = df["Assists"].sum()
print(total_assists)
```

```python
#total penalty goals
total_penalty_goals = df["Penalty_Goals"].sum()
print(total_penalty_goals)
```

```python
#piechart for the penalties scored vs missed
plt.figure(figsize=(7,5))
pl_not_scored = df["Penalty_Attempted"].sum() - total_penalty_goals
data = [pl_not_scored,total_penalty_goals]
labels = [ "missed", "Penalties scored"]
color = sns.color_palette("tab10")
plt.pie(data, labels = labels, colors = color ,autopct= "%.0f%%")
plt.show
```

```python
#lets find their positions like defensive midfielder,striker etc (unique positionm)
df["Position"].unique()
```

```python
#total strikers
df[df["Position"] == "FW"]
```

```python
max_goals_player = df[df['Goals'] == df['Goals'].max()]['Name'].iloc[0]
max_goals = df['Goals'].max()
plt.figure(figsize=(5, 2))
plt.bar(max_goals_player, max_goals, color='blue')
plt.xlabel('Player')
plt.ylabel('Goals')
plt.title('Player with Maximum Goals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.text(max_goals_player, max_goals + 0.5, f'{max_goals}', ha='center')
plt.show()
###here basically u can see tht harry kane was the top scorer from tottenham
```

```python
max_assist_player = df[df['Assists'] == df['Assists'].max()]['Name'].iloc[0]
max_assists = df['Assists'].max()
plt.figure(figsize=(5,2))
plt.bar(max_assist_player, max_assists, color='green')
plt.xlabel('Player')
plt.ylabel('assist')
plt.title('Player with maximum assists')
plt.xticks(rotation=45)
plt.tight_layout()
plt.text(max_assist_player,max_assists  + 0.5, f'{max_assists}', ha='center')
plt.show()
#one season wonder Harry kane also having highest assists
```

```python
#
nationality = df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind= "bar", figsize = (12,5) , color = sns.color_palette("magma"))
plt.title("highest country players", fontsize=15)
#as we can see most of the players are from England (english premier league)
```

```python
#clubs with maximum players intheir squad
df["Club"].value_counts().nlargest(5).plot(kind = "bar" , color = sns.color_palette("magma"))
#ans=westbrom stands around 30 players including reserves
```

```python
#clubs with minimum players
df["Club"].value_counts().nsmallest(5).plot(kind = "bar" , color = sns.color_palette("viridis"))
#as we hav seen leeds united and man city hav the least amount of players
#also we know it will be tough for leeds as marcelo bielsa pressurizes their player for high intensity training

                                                                                
```

```python
under20= df[df["Age"] <= 20]
under20_25= df[(df["Age"] > 20) & (df ["Age"] <= 25)]
under25_30= df[(df["Age"] > 25) & (df ["Age"] <= 30)]
above30= df[df["Age"] > 30]
```

```python
x = np.array([under20['Name'].count(),under20_25["Name"].count(),under25_30["Name"].count(),above30["Name"].count()])
mylabels = ["<=20",">20 & <=25",">25 & <=30", ">30"]
plt.title("Total players with age group", fontsize=15)
plt.pie(x,labels = mylabels , autopct = "%.1f%%")
plt.show()
#here are basically i found that maximum age group is under 20 to 30 its an competitive league
              
```

```python
#most youngsters in a club lets say less then 22 years of age
players_underage_22 = df[df[ "Age"] < 22]
players_underage_22["Club"].value_counts().plot(kind = 'bar', color = sns.color_palette("magma"))
plt.title("age under 22", fontsize=15)
#here basically southampton wolves brighton hav better youngsters 
```

```python
#goals scored by each club for this season
Goals_byclubs = pd.DataFrame(df.groupby("Club" , as_index =False)['Goals'].sum())
sns.set_theme(style="darkgrid",color_codes = True)
ax = sns.barplot(x="Club",y="Goals",data = Goals_byclubs.sort_values(by="Goals"),palette="rocket")
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title("clubs total goals" , fontsize=20)
             
```
